# Process categorical data

```python
categorical_features = [ 'MSZoning', 'Street','LotShape', 'LandContour','Utilities','LotConfig', 'LandSlope','Neighborhood', 'Condition1', 
                        'Condition2', 'BldgType','HouseStyle',  'RoofStyle','RoofMatl', 'Exterior1st','Exterior2nd',  'MasVnrType', 'ExterQual', 'ExterCond',
                        'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 
                        'Electrical', 'KitchenQual', 'Functional', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive',
                        'SaleType','SaleCondition', 'CentralAir'] 
```

```python
len(categorical_features)
```

```python
def categorigal_enc(feature_list, df):
    i = 0
    for feat in feature_list:
        dummy_df = pd.get_dummies(df[feat], drop_first=True)
        df.drop([feat],axis=1,inplace=True)

        if i==0:
            dummy_df_full = dummy_df
        else:
            dummy_df_full = pd.concat( [dummy_df_full, dummy_df], axis=1)

        i+=1

    df_final = pd.concat([df, dummy_df_full] , axis=1)

    return df_final
        
            
            
    
```

```python
df_full = pd.concat([train_df, test_df], axis=0)
```

```python
df_full.info()
```

```python
df_train_test_final=categorigal_enc(categorical_features , df_full)
```

```python
df_train_test_final= df_train_test_final.loc[:, ~df_train_test_final.columns.duplicated()]
```

```python
df_train_test_final.info()
```

```python
train_df = df_train_test_final.iloc[:1459, :]
test_df = df_train_test_final.iloc[1460:, :]
```

## Modelling

```python
X_train = train_df.drop(['SalePrice'], axis=1)
y_train = train_df[['SalePrice']]

```

```python
X_test = test_df.drop(['SalePrice'], axis=1)
```

```python
import xgboost
classifier=xgboost.XGBRegressor()
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
pred = pd.DataFrame(y_pred)

#take id name from kaggle submission file
id_df = pd.read_csv('sample_submission.csv')
ds = pd.concat([id_df['Id'], pred], axis = 1)
ds.columns = ['Id', 'SalePrice']

ds.to_csv('sample_submission_ilora.csv', index = False)
```
